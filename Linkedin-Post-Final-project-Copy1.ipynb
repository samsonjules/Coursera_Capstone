```python

ids,institutions,posts,domaines,date_limites = [],[],[],[],[]
pages = 205
for page in range(1,pages):
    url =f"https://www.jobpaw.com/pont/professionnels.php?pageNum_RS_job={page}&totalRows_RS_job=10622&id=55"
    print(url)
    brower.get(url)
    soup = BeautifulSoup(brower.page_source ,'html5')
    tables = soup.find_all('table',class_='table table-hover table-bordered')
    trs = tbody[0].find_all('tr')
    for tr in trs:
        tds = tr.find_all('td')
        if len(tds) ==4 :
            a = tds[0].find('a')
            id_ =  a.get('href').split('?idj=')[-1]
            ids.append(id_)
            name =  tds[0].get_text().strip()
            post =  tds[1].get_text().strip()
            domaine =  tds[2].get_text().strip()
            date_limit =  tds[3].get_text().strip()
            institutions.append(name)
            posts.append(post)
            domaines.append(domaine)
            date_limites.append(date_limit)
    time.sleep(3)
df_3 = pd.DataFrame(dict(id_=ids,institution=institutions,post=posts,domaine=domaines))
df_3.to_csv('post4.csv')
```
